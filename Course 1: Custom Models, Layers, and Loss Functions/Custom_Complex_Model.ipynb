##Build an Inception-esc Model

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input, concatenate
from tensorflow.keras.models import Model 
from tensorflow.nn import relu
from tensorflow.keras.utils import plot_model
```

##Define Layers 

```python
input_deep = Input(shape=(1,), name='input_deep')
input_wide = Input(shape=(1,), name='input_wide')
hidden_deep_1 = Dense(30, activation='relu', name='hidden_deep_1')(input_deep)
hidden_deep_2 = Dense(30, activation='relu', name='hidden_deep_2')(hidden_deep_1)
concat = concatenate([input_wide, hidden_deep_2], name='concatenate')
output = Dense(1, name='output_unit')(concat)
```

##Build Model

```python
model = Model(inputs = [input_deep, input_wide], 
              outputs = output)

plot_model(model, to_file='model.png', show_shapes=True, show_layer_names=True)
```
