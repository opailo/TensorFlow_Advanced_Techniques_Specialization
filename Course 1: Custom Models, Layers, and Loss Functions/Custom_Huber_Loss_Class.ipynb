```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.losses import Loss

import numpy as np
```

```python
xs = np.arange(-5, 5, 1).astype('float32')
ys = xs * 2 - 1
```

##Build a Class for the Huber Loss 

```python
class Huber(Loss):
  threshold = 1

  def __init__(self, threshold):
    super().__init__()
    self.threshold = threshold
  
  def call(self, y_true, y_pred):
    error = y_true - y_pred

    #Condition that will return r1 if TRUE, and r2 if FALSE
    return_type = tf.abs(error) <= self.threshold

    r1 = 0.5 * tf.square(error)
    r2 = self.threshold * (tf.abs(error) - (0.5 * self.threshold))
    
    return tf.where(return_type, r1, r2)
```

```python
input = Input(shape=(1,), name='input_layer')
output_layer = Dense(1, name='output_layer')(input)

model = tf.keras.models.Model(inputs=input, outputs=output_layer)

#Set parameters
model.compile(optimizer='sgd', loss=Huber(threshold = 1))

model.fit(xs, ys, epochs = 500, verbose = 0)
model.predict([10.0])
```
