```python
import tensorflow as tf
```

##Add function


add `@tf.function` as a wrapper to any function to convert it to Graph code 

```python
@tf.function
def add(a, b):
  return a + b

a = tf.Variable([[1., 2.], [3., 4.]])
b = tf.Variable([[4., 0.], [1., 5.]])

print(tf.add(a, b))

print(tf.autograph.to_code(add.python_function))
```

```python
##Basic `if` function

@tf.function
def f(x):
  if x > 0:
    x = x * x
  return x

print(tf.autograph.to_code(f.python_function))
```
