There are typically three ways of performing style transfer learning with deep learning:

##Supervised Learning 
This involves building a network that is given pairs of iumages and is expected to learn how the style of one should match the contents of the other 
* This requires pairs of images with both the original and the desired style
 * This most likely requires actual artistry to form the two similar images that only differ in style

##Neural Style Transfer 
This is the more common method and is what will be covered. In this method it's not necessary to train a network to understand how one image should map onto another. Instead, the features are extracted using a network which are then applied to an image as styles
* Generally neural style transfer begins with a pre-trained model which can be trained on a single pair of images.
* Use the model to extract style from the first image and to extract the content from the second image. 
* Then create a new image with elements of the features and content from both of the original images.
 * This is performed iteratively and through a loop by minimizing the loss of the generated image with respect to the style of image one and the content of image two
 * Loss minimization is used to merge the style and content into the generated image

##Style Transfer: Conceptual Overview 

image.png

There are three images:
* `content image`
* `style image`
* `generated image`

Use a pre-trained model like `VGG-19` to extract the features of the `content image`
* The features will be stored in the last block of the model where they will be sent to the `content loss` function

Style features will also be extracted using a pre-trained model 
* The main difference is that style features can be extracted from every layer of the model as shown by the arrows leading to the `style loss` function

The `generated image` is initialized through the content image 
* the two's differences can be compared through the content loss
 * Initially, the content loss will be very small, but as the style features get applied to the `generated image` the loss progressively increases 
 * The difference between the `content image` and `generated image` is checked on every iteration 
* The `generated image` amd the `style image` can be compared as well through the `style loss`
 * The style loss is calculated by comparing each layer in the generated image model with the mirrored layer in the style image model. The final `style loss` is an average of all these differences 
 * Similarly, as the style features are applied to the `generated image`, the style loss decreases
* The `style loss` and `content loss` are combined together to form the `total loss`
 * Ideally the `total loss` would be low so that the content is still recognizable but the stylized features are still there
 * An optimizer is used to update the features with respect to the `total loss`. Each upadate results in a new `generated image` that is (hopefully) slightly better than the last.
